```python
from collections import OrderedDict 
import pandas as pd

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
```

```python
df = pd.read_csv("~/Projects/meta_analysis/manual_review/Final Sheets/ridout_annotated.csv")
print(df.head())

abstracts = list(df["Abstract.Note"])
stop_words = set(stopwords.words('english'))
stop_words.add(",")
stop_words.add(".")
stop_words.add(")")
stop_words.add("(")
stop_words.add("'s")
stop_words.add("'")
stop_words.add("%")
stop_words.add(":")
stop_words.add(";")
stop_words.add("?")
res_dict = {}
stres_dict = {}
cond_dict = {}
bio_dict = {}

def update_dict(d, terms):
    for term in terms:
        if term not in d.keys():
            d[term] = 1 
        else:
            d[term] += 1
    return d

def write_dict(d, path):
    file = open(path, "w+")
    file.write("Term, count\n")
    for i in d.keys():
        file.write(i+","+str(d[i])+"\n")
    file.close()
        
for i in range(len(abstracts)):
    filtered_abstract = [w for w in word_tokenize(abstracts[i].lower()) if not w in stop_words] 
    if list(df["Resilience"])[i] == 1:
        res_dict = update_dict(res_dict, filtered_abstract)
    if list(df["Stressors"])[i] == 1:
        stres_dict = update_dict(stres_dict, filtered_abstract)
    if list(df["Biomarkers"])[i] == 1:
        cond_dict = update_dict(cond_dict, filtered_abstract)
    if list(df["Conditions"])[i] == 1:
        bio_dict = update_dict(bio_dict, filtered_abstract)
        
res_dict = {k: v for k, v in sorted(res_dict.items(),reverse=True, key=lambda item: item[1])}
stres_dict = {k: v for k, v in sorted(stres_dict.items(),reverse=True, key=lambda item: item[1])}
cond_dict = {k: v for k, v in sorted(cond_dict.items(),reverse=True, key=lambda item: item[1])}
bio_dict = {k: v for k, v in sorted(bio_dict.items(),reverse=True, key=lambda item: item[1])}


write_dict(res_dict, "res_terms.csv")
write_dict(stres_dict, "strss_terms.csv")
write_dict(cond_dict, "cond_terms.csv")
write_dict(bio_dict, "bio_terms.csv")
```
